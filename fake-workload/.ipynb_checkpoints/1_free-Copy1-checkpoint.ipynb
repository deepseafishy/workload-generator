```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as figure
import json
```

```python
color = {
    '1_free': 'green',
    '4_prod': 'red',
    '2_beb': 'blue',
    '3_mid': 'yellow',
    'default': 'black'
}
```

```python
maxi = 1
for i in range(maxi):
    if i < 10:
        file_name = 'json/data-00000000000{}'.format(i)
    else:
        file_name = 'json/data-0000000000{}'.format(i)
    file = pd.read_csv(file_name)
    
    avg_cpu = []
    avg_mem = []
    time = []
    prev = 0
    prev_prio = 'default'
    max_diff = 0
    maxi = 0
    mini = 1
    for index, row in file.iterrows():
        if(row['tier'] != '1_free'):
            continue;
        row['time'] = row['time']/(1e6*5*60)
        if row['avg_cpu_usage'] > maxi:
            maxi = row['avg_cpu_usage']
        if row['avg_cpu_usage'] < mini:
            mini = row['avg_cpu_usage']
        
        task_id = str(row['machine_id']) +"_"+ str(row['collection_id']) +"_"+ str(row['instance_index'])+"_"
        if task_id != prev:
            if max_diff < maxi-mini:
                max_diff = maxi-mini
            else:
                continue;
            plt.figure(figsize=(15, 6), dpi=80)
            assert(len(time) == len(avg_cpu))
            assert(sorted(time) == time)
            
#             temp_in = open('json/in', 'w+')
#             temp_in.write(json.dumps({'time': time, 
#                                       'avg_cpu': avg_cpu, 
#                                       'avg_mem': avg_mem,
#                                       '' 
#                                      }))

            col = color[prev_prio]
            plt.plot((time), avg_cpu, color=col)
            plt.title(str(prev) + "_" + row['tier'] + "__" + str(index))
#             plt.xlin([0, 8000])
            plt.ylim([mini-0.0005, maxi+0.0005])
            plt.show()
            avg_cpu = []
            avg_mem = []
            time = []
            prev = task_id
            prev_prio = row['tier']
            mini = 1
            maxi = -1
            
        avg_cpu.append(row['avg_cpu_usage'])
        avg_mem.append(row['avg_memory_usage'])
        time.append(row['time'])
    
    
```
